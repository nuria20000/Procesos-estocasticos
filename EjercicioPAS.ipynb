```python
import numpy as np
from matplotlib import pyplot as plt
```

## El núcleo: implementar una función para simular el PAS

Define una función `simular_PAS` que reciba tres argumentos:
* p: este es el parámetro principal del PAS. Es la probabilidad de tomar el valor 1 en cada paso (y la probabilidad de tomar el -1 sería $1-p$). Es un valor entre 0 y 1.
* n: el número de pasos a simular para cada trayectoria. Un número natural.
* reps: el número de trayectorias a simular. Un número natural.

La función debe devolver un array de numpy con dimensionalidad `(reps, n)`, que contenga los valores del PAS en cada paso para cada trayectoria.  

Esqueleto de la función a construir (a completar por vosotros):

```python
def simular_PAS(p,n,reps=1):
  # Input:
  # p: La probabilidad de tomar el valor 1 en cada paso
  # n: Numero de pasos a simular
  # reps: Numero de trayectorias a simular

  # Output:
  pasos=np.random.choice([1,-1], size=(reps,n),p=[p,1-p])  #Aqui tenemos un array de dimension (reps, n), donde cada fila es una trayectoria de n pasos del PAS
  #trayectorias=np.apply_along_axis(np.cumsum,1,pasos) Le aplica a cada fila la funcion cumsum, de tal forma que nos da, las trayectorias en cada fila
  trayectorias = np.cumsum(pasos, axis=1) #que hace que la funcion se aplica por filas

  #print("Tamaño del array de los pasos" ,pasos.shape)
  #print("Tamaño del array de las trayectorias", trayectorias.shape)
  #print("Array de las trayectorias",trayectorias) En vez de poner el return trayectorias tambien podriamos haber usado esto


  return trayectorias  # Shape: (reps, n)
```

## Visualización 1: diferentes trayectorias del PAS (sin media ni desviación típica teóricas)

PAS con p=0.5

```python

datos=simular_PAS(0.5, 500, 20)
t = np.arange(1, datos.shape[1] + 1)
plt.plot(t,datos.T)

plt.title("20 trayectorias de un PAS con p=0.5")
plt.xlabel('t')
plt.ylabel("$X_t$")
plt.grid(True)
plt.show()
```

## Visualización 2: diferentes trayectorias del PAS (con media y desviación típica teóricas)

PAS con p=0.5

```python
datos=simular_PAS(0.5, 500, 20)
t = np.arange(1, datos.shape[1] + 1)
plt.plot(t,datos.T)

p=0.5
media_teorica = (2*p - 1) * t  #Media teorica
plt.plot(t,media_teorica,color='k',linestyle='--', label="Media teorica")
sigma=np.sqrt(4*p*(1-p)*t) #Desviacion tipica

plt.fill_between(t, media_teorica + 2*sigma, media_teorica - 2*sigma, color='gray', alpha=0.15,label=r"$\pm 2\sigma$") #Sombreado grande ,se pone la r, para que o de error
plt.fill_between(t, media_teorica + sigma, media_teorica -sigma, color='gray', alpha=0.3,label=r"$\pm 1\sigma$") #Sombreado pequeño


plt.title("20 trayectorias de un PAS con p=0.5")
plt.xlabel('t')
plt.ylabel("$X_t$")
plt.grid(True)
plt.legend(loc='upper left', frameon=True, shadow=True);
```

PAS con p=0.75

```python
datos=simular_PAS(0.75, 500, 20)
t = np.arange(1, datos.shape[1] + 1)
plt.plot(t,datos.T)

p=0.75
media_teorica = (2*p - 1) * t  #Media teorica 2p-1
plt.plot(t,media_teorica,color='k',linestyle='--', label="Media teorica")

sigma=np.sqrt(4*p*(1-p)*t) #Desviacion tipica 4p(1-p)
plt.fill_between(t, media_teorica + 2*sigma, media_teorica - 2*sigma, color='gray', alpha=0.15,label=r"$\pm 2\sigma$") #Sombreado grande ,se pone la r, para que o de error
plt.fill_between(t, media_teorica + sigma, media_teorica -sigma, color='gray', alpha=0.3,label=r"$\pm 1\sigma$") #Sombreado pequeño


plt.title("20 trayectorias de un PAS con p=0.75")
plt.xlabel('t')
plt.ylabel("$X_t$")
plt.grid(True)
plt.legend(loc='upper left', frameon=True, shadow=True);
```

PAS con 0.95

```python
datos=simular_PAS(0.95, 500, 20)
t = np.arange(1, datos.shape[1] + 1)
plt.plot(t,datos.T)

p=0.95
media_teorica = (2*p - 1) * t  #Media teorica 2p-1
plt.plot(t,media_teorica,color='k',linestyle='--', label="Media teorica")

sigma=np.sqrt(4*p*(1-p)*t) #Desviacion tipica 4p(1-p)
plt.fill_between(t, media_teorica + 2*sigma, media_teorica - 2*sigma, color='gray', alpha=0.15,label=r"$\pm 2\sigma$") #Sombreado grande ,se pone la r, para que o de error
plt.fill_between(t, media_teorica + sigma, media_teorica -sigma, color='gray', alpha=0.3,label=r"$\pm 1\sigma$") #Sombreado pequeño



plt.title("20 trayectorias de un PAS con p=0.95")
plt.xlabel('t')
plt.ylabel("$X_t$")
plt.grid(True)
plt.legend(loc='upper left', frameon=True, shadow=True);
```

## Visualización 3: visualización de la media empírica del PAS cuando aumentamos el número de trayectorias (junto con la media teórica)

```python
p=0.5
cantidades = [5, 100, 200, 500, 1000] #El numero de veces que se van a repetir las trayectorias, estan puestas en la diapositiva

for i, reps in enumerate(cantidades):   #Aqui lo que hacemos es un bucle, para que calcule la media de cada trayectoria segun el numero de repeticiones que ha hecho
 trayectorias = simular_PAS(p, 500, reps)
 media_empirica = np.mean(trayectorias, axis=0)  #Axis=0 para que le haga la media a las filas
 plt.plot(t, media_empirica, label=f'{reps}')


t = np.arange(500)
media_teorica = (2*p - 1) * t  #Media teorica
plt.plot(t,media_teorica,color='black', linewidth=2, label='Exact')

plt.xlabel("t")
plt.ylabel("$X_t$")
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```
